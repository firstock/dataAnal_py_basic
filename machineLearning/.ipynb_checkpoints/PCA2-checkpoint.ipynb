```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_breast_cancer
cancer= load_breast_cancer()
```

```python
from sklearn.preprocessing import StandardScaler
```

##  PCA+ StandardScaler로 스케일조정

```python
scaler= StandardScaler()
scaler.fit(cancer.data)
```

```python
X_scaled= scaler.transform(cancer.data)
```

```python
X_scaled
```

##### PCA: 주성분 2개

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components= 2)
```

```python
pca
```

유방암데이터로 PCA모델 만들기

```python
pca.fit(X_scaled)
```

주성분 2개로 데이터 변환

```python
X_pca= pca.transform(X_scaled)
```

```python
print('원본형태',str(X_scaled.shape))
print('축소형태',str(X_pca.shape))
```

### 시각화: 주성분 2개로 산점도
.discrete_scatter 3번째인자: 클래스별로 다른 색

```python
!pip install mglearn
```

```python
import matplotlib.pyplot as plt
import mglearn
%matplotlib inline
```

```python
plt.figure(figsize=(8,8))
mglearn.discrete_scatter(X_pca[:,0], X_pca[:,1], cancer.target)

font={
    'family':'Malgun Gothic'
    ,'size':15
}
plt.rc('font', **font)
plt.legend(['악성','양성'], loc= 'best')
plt.gca().set_aspect('equal')

plt.xlabel('첫 번째 주성분')
plt.ylabel('두 번째 주성분')
```

### 주성분 형태 확인
PCA(...).components_ : 주성분을 구성하는 각 계수들

```python
print('PCA주성분 형태:', pca.components_.shape)
```

(0.2189)A1 , (-0.23385)B1 + ... + (0.13178)A30 , (0.27533)B30

```python
print('PCA주성분: ', pca.components_)
```

```python
print('클래스 개수:',len(people.target_names))
people.images.shape
```

#### 각 클래스 빈도 확인
각 target이 나타난 횟수 계산, target별 이름과 횟수 출력

```python
import numpy as np
```

```python
counts= np.bincount(people.target)
```

```python
for i, (count, name) in enumerate(zip(counts, people.target_names)):
    print('{0:25} {1:3}'.format(name,count), end='\t')
    if (i+1)%3== 0:
        print()
```

#### 사람별 50개씩만 선택- 데이터 편중 방지

```python
mask= np.zeros(people.target.shape, dtype= np.bool)
```

```python
for target in np.unique(people.target):
    mask[np.where(people.target== target)[0][:50]]= 1
```

```python
X_people= people.data[mask]
y_people= people.target[mask]
```

```python
len(X_people), len(X_people[0])
```

```python
len(y_people)
```

0~255 /255 = 0~1 . MinMaxScaler나 마찬가진데, 속도좀더 빠르게

```python
X_people= X_people/255.
```

### K-NN 거리계산으로 img class 분류
얼굴 유사도- 같은 픽셀위치에 있는거랑 값 크기 비교(거리를 계산)
> 고개를 돌리고 있거나, 여백만 좀 있어도 확 낮아지는 score

부정확해서 이제는 잘 안쓰는 방법인데, 비교하려고 해봄

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test= train_test_split(X_people, y_people, stratify= y_people, random_state= 0)
```

k= 1

```python
knn= KNeighborsClassifier(n_neighbors= 1)
```

```python
knn.fit(X_train, y_train)
```

```python
print('1-NN test score%9.5f'%knn.score(X_test, y_test))
```

#### 0.232 ??? 값이 무척 낮지만, 얼굴인식 분야에서 이정도면 '괜찮다' 고 평함
아직 최적의 방법이 안 나와서

## PCA변환 -> KNN
주성분으로 변환후 거리 계산

###### ? PCA모델(화이트닝??옵션= True) == 스케일(StandardScaler) 
PCA객체~ train data 학습, 설명력 높은 순으로 100개의 주성분 추출

```python
pca= PCA(n_components= 100, whiten= True, random_state= 0).fit(X_train)
```

```python
pca
```

```python
X_train_pca= pca.transform(X_train)
X_test_pca= pca.transform(X_test)
```

```python
type(X_train_pca)
```

```python
X_train_pca.shape
```

```python
X_test_pca.shape
```

### K-NN으로 분류분석
변수도 깔쌈하게 합쳐놨으니, 거리계산을 해보자

```python
knn= KNeighborsClassifier(n_neighbors= 1)
knn.fit(X_train_pca, y_train)
```

```python
print('test:%9.5f'%knn.score(X_test_pca, y_test))
```

#### K-NN단독으로 돌려서 나온 0.232 보다 0.08 올라간 0.312

```python
0.31202-0.23256
```

```python
fig, axis= plt.subplots(3,5, figsize= (15,12)
                       , subplot_kw= {'xticks':(), 'yticks':()})
```

```python
for i, (component, ax) in enumerate(zip(pca.components_, axis.ravel())):
    ax.imshow(component.reshape(image_shape), cmap= 'inferno')#viridis
    ax.set_title('주성분%2d'%(i+1))
```
