### 고차원 데이터 시각화

```python
import matplotlib.pyplot as plt
import mglearn
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
cancer= load_breast_cancer()
```

```python
import numpy as np
```

```python
# print('평탄',len(axis.ravel()))
# print('그냥axis',len(axis))
```

```python
fig.clear('all')
```

```python
fig, axis= plt.subplots(15,2, figsize=(10,20))

#cancer.data[cancer.target== 0] 형태: (212,30)
#cancer.data[cancer.target== 1] 형태: (357,30)
malignant= cancer.data[cancer.target== 0]
benign= cancer.data[cancer.target== 1]

ax= axis.ravel() #평탄화

for i in range(30):
    _, bins= np.histogram(cancer.data[:,i], bins=50)
    
    ax[i].hist(malignant[:,i], bins=bins, color= mglearn.cm3(0), alpha=.5) #보라색
    ax[i].hist(benign[:,i], bins=bins, color= mglearn.cm3(2), alpha=.5) #초록색, 연두색
    
    ax[i].set_title(cancer.feature_names[i])
    ax[i].set_yticks(())
    
ax[0].set_xlabel('특성 크기')
ax[0].legend(['악성','양성'], loc='best')
fig.tight_layout()
```
