### 함수 정의부
#### 이미지 출력을 함수화

```python
# filePath= "data/tree.raw"
def load_img(filePath):
    """
    RAW file --> memory(2 dimensional list)
    param(str filePath/fileName)
    return(2D list)
    """
    inImage= []
    try:
        with open(filePath,'rb') as fp:
            for i in range(0,XSIZE):
                tmpList= []
                for k in range(0,YSIZE):
                    data= ord(fp.read(1))
                    tmpList.append(data)
                inImage.append(tmpList)
    except FileNotFoundError:
        print("load_img: 해당 파일이 없습니다")
        return ""
    else:
        print("success!- load on memory")
        return inImage
```

```python
def display_img(imgL2):
    """
    param(fuction load_img)
    """
    try:
        for i in range(0,XSIZE):
            for k in range(0,YSIZE):
                data= imgL2[i][k]
                paper.put("#%02x%02x%02x"%(data,data,data),(k,i))
    except :
        print("display_img: 바른 인자 형식이 아닙니다> load_img - FileNotFoundError")
```
