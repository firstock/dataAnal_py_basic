## 주의점- 상관계수??
예시는 변화가 일관되게 상승하는 두 함수이고,
공정 샘플은 상승했다 하강하는 데이터라, 구간 나누지 않고 상관계수보는 건 위험

### 예시
[dtw](https://yjucho1.github.io/time-series/dtw/?fbclid=IwAR1iqKbUGkorULDTOal4F05tZTqM0DEgV6xoOpTAtJ8aOa6IxN6bP7oW-Dc)

```python
ts1 = [1, 5, 3, 4, 7, 6]
ts2 = [0, 2, 6, 3, 5, 6, 8, 5]
```

```python
x = np.array(ts1).reshape(-1,1)
y = np.array(ts2).reshape(-1,1)

euclidean_norm = lambda x, y: np.abs(x - y)
```

```python
type(euclidean_norm)
```

```python
# dist = type(function)
d, cost_matrix, acc_cost_matrix, path = dtw(x, y, dist=euclidean_norm)
```

```python
type(ts1)
```

```python
plt.imshow(acc_cost_matrix.T, origin='lower', cmap='gray', interpolation='nearest')
plt.plot(path[0], path[1], 'w')
plt.show()
```
