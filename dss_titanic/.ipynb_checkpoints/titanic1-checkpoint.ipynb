```python
import pandas as pd
```

## Load Dataset

```python
train= pd.read_csv("data/train.csv",index_col="PassengerId")
print(train.shape)
train.head(2)
```

```python
test= pd.read_csv("data/test.csv",index_col="PassengerId")
print(test.shape)
test.head(2)
```

## Preprocessing


## Encode Sex

```python
train.loc[train["Sex"]=="male","Sex"] =0
train.loc[train["Sex"]=="female","Sex"] =1
```

```python
print(train.shape)
train[["Sex"]].head(5)
```

```python
test.loc[test["Sex"]=="male","Sex"] =0
test.loc[test["Sex"]=="female","Sex"] =1
```

```python
print(test.shape)
test[["Sex"]].head(5)
```

```python
train[train["Fare"].isnull()]
```

```python
test[test["Fare"].isnull()]
```

```python
test.loc[test["Fare"].isnull(),"Fare"]=0
test[test["Fare"].isnull()]
```

## Encode Embarked

## one-hot encoding

```python
train["Embarked_C"]= train["Embarked"]=="C"
train["Embarked_S"]= train["Embarked"]=="S"
train["Embarked_Q"]= train["Embarked"]=="Q"

print(train.shape)
train[["Embarked_C","Embarked_S","Embarked_Q"]].head(4)
```

```python
test["Embarked_C"]= test["Embarked"]=="C"
test["Embarked_S"]= test["Embarked"]=="S"
test["Embarked_Q"]= test["Embarked"]=="Q"
```

```python
print(test.shape)

train[["Embarked_C","Embarked_S","Embarked_Q"]]
```

```python
def testOneHot(colName1,colName2,colName3):
    test["Embarked_"+colName1]= test["Embarked"]==colName1
    test["Embarked_"+colName2]= test["Embarked"]==colName2
    test["Embarked_"+colName3]= test["Embarked"]==colName3
    test[["Embarked_C","Embarked_S","Embarked_Q"]].head(4)    

testOneHot("C","S","Q")
print(test.shape)

train[["Embarked_C","Embarked_S","Embarked_Q"]]
```

### Train

```python
feature_names= ["Pclass","Sex","Fare","Embarked_C","Embarked_S","Embarked_Q"]
feature_names
```

```python
X_train= train[feature_names]
print(X_train.shape)
X_train.head()
```

```python
X_test= test[feature_names]
print(X_test.shape)
X_test.head()
```

```python
label_name= "Survived"
y_train= train[label_name]

print(y_train.shape)
y_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier(max_depth=5)
model
```

```python
model.fit(X_train, y_train)
```

```python
predictions= model.predict(X_test)
print(predictions.shape)
predictions[0:10]
```

## submit

```python
submit= pd.read_csv("data/gender_submission.csv",index_col="PassengerId")

print(submit.shape)
submit.head()
```
